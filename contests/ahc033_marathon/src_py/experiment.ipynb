```python
import random
for (sample_n, A) in enumerate(inputs):
    for try_n in range(100):
        order = []
        for i in range(5):
            order += [i]*5
        random.shuffle(order)
        okiba = []
        A_ = [[A[i][4-j] for j in range(5)] for i in range(5)]
        okiba = []
        cur = [i*5 for i in range(5)]
        is_ok = True
        for idx in order:
            flg = True
            while flg:
                flg = False
                for i in range(5):
                    for i_ in range(5):
                        if A_[i] and A_[i][-1]==cur[i_]:
                            cur[i_] += 1
                            flg = True
                            A_[i].pop()
                            break
                for i_ in range(5):
                    if cur[i_] in okiba:
                        okiba.remove(cur[i_])
                        flg = True
                        cur[i_] += 1
                        break
            if A_[idx]:
                okiba.append(A_[idx].pop())
            if len(okiba) > 8:
                is_ok = False
                break
        if is_ok:
            # print(f"sample {sample_n} is ok in {try_n}")
            break
    if not is_ok:
        print(f"sample {sample_n} is not ok")
        break
```

```python
import ast
import matplotlib.pyplot as plt
import numpy as np

arrow_dict = {
    0: (0, -0.3),
    1: (0.3, 0),
    2: (0, 0.3),
    3: (-0.3, 0),
    4: "P",
    5: "S",
}

colors = ["red", "orange", "green", "blue", "blask"]

def visualize_map(s, title):
    s = s.replace("Map2d::new(", "").replace(")", "").replace("vec!", "").replace(";", "").replace("(", "")
    M = ast.literal_eval(s)
    # グリッドサイズを取得
    rows = 5
    cols = 5

    # プロットの準備
    fig, ax = plt.subplots()

    # グリッドの設定
    ax.set_xticks(np.arange(0, cols, 1))
    ax.set_yticks(np.arange(0, rows, 1))
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.grid(which='both')
    ax.set_xlim(-0.5, cols-0.5)
    ax.set_ylim(-0.5, rows-0.5)
    ax.set_title(title)

    # 矢印の描画
    for i in range(rows):
        for j in range(cols):
            for color_idx, arrow in enumerate(M[i*5+j]):
                if arrow>=4:
                    ax.text(j, i, f"{arrow_dict[arrow]}", ha="center", va="center", color=colors[color_idx])
                else:
                    dx, dy = arrow_dict.get(arrow, (0, 0))
                    ax.arrow(j, i, dx, dy, head_width=0.1, head_length=0.1, fc=colors[color_idx], ec=colors[color_idx])

    plt.gca().invert_yaxis()
    plt.savefig(f"../vis/map_{title}.png")
    plt.show()
    return M

for map_string in map_strings.split(";"):
    title = map_string.split(":")[0].split("\n")[-1]
    s = map_string.split("=")[1].strip()
    visualize_map(s, title)
```
