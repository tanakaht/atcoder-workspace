```python
def const_S(N):
    S = set()
    apperaed = set()
    while len(S)<N:
        for i in range(10000):
            if i in S:
                continue
            found = False
            for x in S:
                if abs(i-x) in apperaed:
                    found = True
                    break
            if not found:
                for x in S:
                    apperaed.add(abs(i-x))
                S.add(i)
                break
    return sorted(S)

for i in range(30):
    x  = const_S(i)
    print(i, [x[j+1]-x[j] for j in range(i-1)], x)
```

```python
# xこの要素を持つsを作る
def construct(x):
    ret = [0]
    for i in range(len(bin(x))):
        # 2倍にする
        ret2 = [x+2*ret[-1]+1 for x in ret]
        ret += ret2
    ret = ret[:x]
    return ret

construct(10000)[-1]
```
