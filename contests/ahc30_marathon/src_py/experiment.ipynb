```python
import math
import numpy as np

rng = np.random.default_rng()
def ptn(k, v, e, n):
    return sorted([rng.normal((k-v)*e+v*(1-e), math.sqrt(k*e*(1-e))) for _ in range(n)])
print(ptn(9, 1, 0.2, 100))
```

```python
import numpy as np
from scipy.stats import norm #normは正規分布
import matplotlib.pyplot as plt
import math

def draw(k, e, n):
    X = np.arange(-1, 2, 0.01)
    #pdfは確率密度関数
    sigma2 = k*e*(1-e)/math.sqrt(n)
    #平均:60, 標準偏差:10
    Y = norm.pdf(X, 1, math.sqrt(sigma2))
    #グラフの表示
    plt.plot(X, Y)
    plt.show()
draw(9, 0.2, 1)
```
