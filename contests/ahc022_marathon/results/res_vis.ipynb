```python
import sys
import matplotlib.pyplot as plt
import sys
d = {}
en = 99
for i in range(en+1):
    with open(f"../tools/err/{i:04}.txt", "r") as f:
        d[i] = {}
        for line in f:
            if line.startswith("Score ="):
                d[i]["score"] = int(line.split()[-1])
            elif line.startswith("Number of wrong answers ="):
                d[i]["n_wrong"] = int(line.split()[-1])
            elif line.startswith("Placement cost ="):
                d[i]["P_cost"] = int(line.split()[-1])
            elif line.startswith("Measurement cost ="):
                d[i]["M_cost"] = int(line.split()[-1])
            elif line.startswith("Measurement count ="):
                d[i]["M_cnt"] = int(line.split()[-1])
            elif line.startswith("x ="):
                d[i]["x"] = int(line.split()[-1])
            elif line.startswith("Number of overlap = "):
                d[i]["n_overlap_init"] = int(line.split()[-1])
            elif line.startswith("Number of overlap final = "):
                d[i]["n_overlap_final"] = int(line.split()[-1])
            elif line.startswith("n planed measure = "):
                d[i]["n_planed_measure"] = int(line.split()[-1])

scores = [d[i]["score"] for i in range(len(d))]
plt.plot(range(len(scores)), sorted([math.log(x, 10) for x in scores]))

```

```python
for i in range(100):
    try:
        d[i]["score"]
    except KeyError:
        print(i)
```

```python
def get_k_v(k):
    return [d[i][k] for i in range(len(d))]
```

```python
import math
eles = [math.log(k/v, 10) for k, v, x in sorted(zip(get_k_v("P_cost"), get_k_v("M_cost"), get_k_v("x")), key=lambda x: x[2])]
xs = [x for k, v, x in sorted(zip(get_k_v("P_cost"), get_k_v("M_cost"), get_k_v("x")), key=lambda x: x[2])]
vs = [v for k, v, x in sorted(zip(get_k_v("P_cost"), get_k_v("M_cnt"), get_k_v("x")), key=lambda x: x[2])]

plt.plot(range(len(eles)), eles)
plt.show()
plt.plot(range(len(xs)), xs)
plt.show()
plt.plot(range(len(xs)), vs)
plt.show()
```
