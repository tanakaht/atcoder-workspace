```python
inputs = []
n_bags = []
for i in range(1000):
    with open(f"../tools/in/{i:04}.txt") as f:
        M, W, H, B, D= map(int, f.readline().split())
        bags = [list(f.readline().split()) for _ in range(M)]
        inputs.append((M, W, H, B, D, bags))
        n_bag = 0
        for h, w, d, a, f, g in bags:
            n_bag += int(a)
        n_bags.append(n_bag)

```

```python
g_sums = []
for M, W, H, B, D, bags in inputs:
    g_sum = 0
    for h, w, d, a, f, g in bags:
        if g=="N":
            if f=="Y":
                g_sum += int(h)*int(w)*int(a)
            else:
                xs = sorted([int(h), int(w), int(d)])
                g_sum += xs[0]*xs[1]*int(a)
    g_sums.append(g_sum)
plt.hist(g_sums, bins=1000)
```
